# **Comparacion inicial de 30 modelos en su forma básica**

## **Importación de librerías necesarias**

```python
# Importando librerías
import pandas as pd
import lazypredict
from lazypredict.Supervised import LazyClassifier
from sklearn.metrics import matthews_corrcoef
```

## **Entrenamiento y predicción de los modelos**

```python
# Definición y construcción de LazyClassifier
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=matthews_corrcoef)

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
X_train = train.drop('class', axis=1)
y_train = train['class'].copy()
X_test = test.drop('class', axis=1)
y_test = test['class'].copy()
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```

# **Métricas de los modelos**

```python
# Plot de la precision
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 10))
sns.set_theme(style="whitegrid")
ax = sns.barplot(y=models.index, x="Accuracy", data=models)
ax.set(xlim=(0, 1))
```

```python
# Plot del coeficiente de correlacion de Mathews
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 10))
sns.set_theme(style="whitegrid")
ax = sns.barplot(y=models.index, x="matthews_corrcoef", data=models)
ax.set(xlim=(0, 1))
```
